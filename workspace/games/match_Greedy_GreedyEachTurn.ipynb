```python
import sys
import os
import pprint

sys.path.append(os.path.join("..", "players"))

from pyrat import Game, PlayerSkin, StartingLocation
from Greedy import Greedy
from GreedyEachTurn import GreedyEachTurn
```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">CONSTANTS</h1>

Let's configure the game with a dictionary.

```python
# Customize the game elements
CONFIG = {"mud_percentage": 20,
          "nb_cheese": 9    ,
          "random_seed": 42,
          "trace_length": 1000}
```

We perform a single game with the configuration defined above.

```python
stock = {"G": 0, "GET": 0}
print("start")
nb_match = 3
for i in range(3):
    # Instantiate a game with specified arguments
    game = Game(**CONFIG)

    # Instantiate a player and add it to the game
    print("exhaustive")
    playerG = Greedy(skin=PlayerSkin.MARIO)  
    playerGET = GreedyEachTurn(skin=PlayerSkin.GHOST)
    print("add_player")
    game.add_player(playerG,team="Greedy", location=StartingLocation.RANDOM)
    game.add_player(playerGET, team="GreedyEachTurn", location=StartingLocation.RANDOM)

    # Start the game and
    stats = game.start()
    if stats['players']['Greedy']['score'] > stats['players']['GreedyEachTurn']['score']:
        stock["G"] += 1
    else:
        stock["GET"] += 1
    victoires_moy = stock["G"] + stock["GET"] / nb_match
    vainqueur_moyen = f"Le vainqueur en moyenne sur 100 matchs est Greedy avec {nb_match} en moyenne" if stock["G"] > stock["GET"] else f"Le vainqueur en moyenne sur 100 matchs est GreedyEachTurn avec {nb_match} en moyenne"
```

```python
# Show statistics
pprint.pprint(stats)
print("\n", vainqueur_moyen)
```
