In this script, we run player `Dijkstra` in a maze, to visualize its behavior.

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">IMPORTS</h1>

```python
import sys
import os
import pprint

sys.path.append(os.path.join("..", "players"))

from pyrat import Game
from GreedyEachCheese import GreedyEachCheese
from Greedy import Greedy
from Exhaustive import Exhaustive
from GreedyEachTurn import GreedyEachTurn
```

<h1 style="background-color: gray;
           color: black;
           padding: 20px;
           text-align: center;">CONSTANTS</h1>

Let's configure the game with a dictionary.

```python
# Customize the game elements
CONFIG = {"mud_percentage": 20,
          "nb_cheese": 9    ,
          "random_seed": 42,
          "trace_length": 1000}
```

We perform a single game with the configuration defined above.

```python
# Instantiate a game with specified arguments
game = Game(**CONFIG)

# Instantiate a player and add it to the game
print("exhaustive")
playerGEC = GreedyEachCheese()
playerG = Greedy()
playerE = Exhaustive()
playerGET = GreedyEachTurn()
print("add_player")
game.add_player(playerGEC, "PlayerGEC")
game.add_player(playerG, "PlayerG")
game.add_player(playerE, "PlayerE")
game.add_player(playerGET, "PlayerGET")

# Start the game and
print("start")
stats = game.start()
```

```python
# Show statistics
pprint.pprint(stats)
```
